```python
#reading the data set
import pandas as pd
filename =  "/kaggle/input/carsforsale/cars_raw.csv"
df = pd.read_csv(filename)
df.head()


```

```python
#by using the info method, i can see the type of data type which i can use later on
df.info()
```

```python
#here i am taking out the discrepencies since "$" makes the column count as a categorical, 
#we need to make it a integer column. In addtion some prices are marked as "Not Priced" which need to be removed

df = df[df['Price'] != 'Not Priced']
df['Price'] = df.Price.str.replace("$", " ")
df['Price'] = df['Price'].astype(int)
```

```python
#changing the 

df['Used/New'] = df['Used/New'].apply(lambda x: 'Certified' if 'Certified' in x else x)

```

```python
#this evens out the mis matched Drivetrains

df = pd.read_csv(filename)
df = df.loc[df['Drivetrain'] != '–']
def drive(x):
    if x == 'Front-wheel Drive' or x == 'Front Wheel Drive' or x == 'FWD':
        return 'FWD'
    if x == 'Four-wheel Drive' or x == 'All-wheel Drive' or x == '4WD' or x == 'AWD':
        return 'AWD'
    if x == 'Rear-wheel Drive' or x == 'RWD':
        return 'RWD'
df['Drivetrain'] = df['Drivetrain'].apply(drive)


df.Drivetrain
```

```python
df.drop(df[df["Drivetrain"]=='–'].index, inplace=True)
```

```python
#generate graphs that shows the different attributes of each car make over the course of time and in the end 
#how consumer rating has changed through the years
import matplotlib.pyplot as plt
variables = ["Make", "Used/New", "Drivetrain", "MinMPG", "MaxMPG", "FuelType", "ConsumerRating" ] 
for var in variables:
    plt.figure(figsize=(14,10))
    plt.title("Over the years" + var + " Shown ")
    uniq = pd.unique(df[var].values.ravel())
    for varr in uniq:
        year = df.loc[df[var] == varr]["Year"]
        varrb = df.loc[df[var] == varr][var]
        plt.plot(varrb, year, label = varr)
    ax = plt.gca()
   
    plt.ylabel('Year')
    plt.legend()
```
